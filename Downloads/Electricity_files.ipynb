```python
import csv
import os
import pandas as pd
```

```python
new_list = []
fwd = list(os.walk('./'))
for file in fwd[0][2]:
    if file.endswith('.csv'):
        with open(file, newline='') as f:
            reader = csv.reader(f)
            next(reader, None)
            data = list(reader)
            new_list.append(data)
```

```python
len(new_list)
```

```python
new_list[2]
```

```python
file = open('test.csv', 'w+', newline ='') 
with file:     
    write = csv.writer(file) 
    for i in new_list:
        write.writerows(i)
```
