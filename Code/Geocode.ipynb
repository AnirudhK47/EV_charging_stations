```python
import pandas as pd
```

```python
df = pd.read_csv("Substation_list_Delhi.csv")
```

```python
df['location'] = df['Substations'] + ', ' + df['Address']
```

```python
import geopy
```

```python
from geopy.geocoders import Nominatim
nom = Nominatim(user_agent="anirudhkaushik1997@gmail.com")
```

```python
df["Latitude"]=df["Coordinates"].apply(lambda x: x.latitude if x!=None else None)
df["Longitude"]=df["Coordinates"].apply(lambda x: x.longitude if x!=None else None)
```
